# Ciclo prob

Avaliação da disponibilidade de um sistema de 2 bombas com regime de alternância pré fixado (0/100 até 50/50). 

A proporção de utilização de cada bomba é alterada para o complementar cada fez que esta é reparada. 

A durabilidade de uma bomba é medida em número de ativações. É criado um vetor aleatório para cada bomba com o número de ativações até falha, que posteriormente é traduzido em tempo, consoante a proporção de funcionamento ativa (que depende se a bomba está em regime forte, fraco, total ou nulo). 

```python
meantime=[]
for lam in range(20,80):   
    n = (2,100000) # number of samples
    k = 3.5     # shape
   # lam = par   # scale

    x = 100*weibull_min.rvs(k, loc=30, scale=lam, size=n)
    #x=np.reshape(x,(2,1000))
    x
    x=np.round(x).astype(int)
    
    #print(x)
    val=np.mean(x)/100
    #np.shape(x)
    meantime.append(val) # tempo médio até à falha
```

```python
for j in range(9):
    #graf=np.zeros((50,2))
    for i in range(50):
        n = (2,5000000) # number of samples
        k = 3.5     # shape
        lam =80-60*((i-50)/50)**2    # scale
        x = 100*weibull_min.rvs(k, loc=30, scale=lam, size=n)
        #x=np.reshape(x,(2,1000))
        x
        x=np.round(x).astype(int)
        prob=i/100
        a=100 # ativações diárias =100
        #prob=0.2 # splitter de probabilidade (valor <=0.5)
        p=prob # proporção de funcionamento da bomba 1
        q=1-prob 
        s1=0 # stock de ativações até avaria da bomba 1
        s2=0
        d=0 # número do dia
        #R=30 # número de dias para a reparação
        dias=5000000
        index1=0
        index2=0
        s1=x[0,0]
        s2=x[1,0]
        lista=[]

        ns1=0
        ns2=0
        while d<dias:

            if s1<0:
                for count in range(ns1,round(np.random.normal(10+5*j,2))):
                    s2=s2-a
                    ns2+=(s2<0)
                    d+=1
                    #print(d,s1,s2)
                    lista.append([d,s1,s2])
                index1+=1
                s1=x[0,index1]
                ns1=0     
                q=1-prob
                p=prob
            if s2<0:
                for count in range(ns2,round(np.random.normal(10+5*j,2))):
                    s1=s1-a
                    ns1+=(s1<0)
                    d+=1
                  #  print(d,s1,s2) 
                    lista.append([d,s1,s2])
                index2+=1
                s2=x[1,index2]
                ns2=0
                q=prob
                p=1-prob
            if s1>=0 and s2>=0:
                    s1=s1-a*p
                    s2=s2-a*q
                 #   print(d,s1,s2)
                    lista.append([d,s1,s2])
                    d+=1
        lista=np.asarray(lista)
        lista_down=(lista[:,1] < 0)  & (lista[:,2] < 0)
        k=np.count_nonzero(lista_down)/dias
        grafs[i,j+1]=k
        #graf=np.asarray(graf)
        #grafs[:,j+1]=graf[:,1]
```

# Modos de falha com preventivas

```python
pp2=3/5
pp3=2/5 #proporções
k=3.5
lam=40
qtl=0.2 #quantile
n=10000 #samples
#cut=200000
```

```python
tq1=weibull_min.ppf(qtl, k, loc=0, scale=lam)
tq1
```

```python
tq2=weibull_min.ppf(qtl, k, loc=0, scale=lam/pp2)
tq2
```

```python
tq3=weibull_min.ppf(qtl, k, loc=0, scale=lam/pp3)
tq3
```

```python
cut=200000
```

```python
k1 = k     # shape
lam1 = lam   # scale

x1 = weibull_min.rvs(k1, loc=0, scale=lam1, size=n)
x1=np.round(x1)

print(x1[:20])
min1=np.minimum(x1,tq1)
print(min1)

sum1=np.cumsum(min1)
sum1=sum1[np.where(sum1<cut)]
min1=min1[:len(sum1)]
tc1=sum1[np.where(min1<tq1)]
print(tc1[:20])
tp1=sum1[np.where(min1==tq1)]
print(tp1[:20])

c1=np.ones(len(tc1))
s1=np.vstack((tc1,c1)).T
s1
print(len(s1))

cp1=np.ones(len(tp1))
sp1=np.vstack((tp1,cp1)).T
print(len(sp1))
```

```python
k2 = k    # shape
lam2 = lam/pp2  # scale

x2 = weibull_min.rvs(k2, loc=0, scale=lam2, size=n)
x2=np.round(x2)

min2=np.minimum(x2,tq2)

sum2=np.cumsum(min2)
sum2=sum2[np.where(sum2<cut)]
min2=min2[:len(sum2)]
tc2=sum2[np.where(min2<tq2)]
tp2=sum2[np.where(min2==tq2)]
c2=np.ones(len(tc2))*2
s2=np.vstack((tc2,c2)).T
s2
print(len(s2))

cp2=np.ones(len(tp2))*2
sp2=np.vstack((tp2,cp2)).T
print(len(sp2))
sp2[:20]
```

```python
#n = 1000 # number of samples
k3 = k    # shape
lam3 = lam/pp3   # scale

x3 = weibull_min.rvs(k3, loc=0, scale=lam3, size=n)
x3=np.round(x3)

min3=np.minimum(x3,tq3)


sum3=np.cumsum(min3)
sum3=sum3[np.where(sum3<cut)]
min3=min3[:len(sum3)]
tc3=sum3[np.where(min3<tq3)]
tp3=sum3[np.where(min3==tq3)]
c3=np.ones(len(tc3))*3
s3=np.vstack((tc3,c3)).T
s3
print(len(s3))

cp3=np.ones(len(tp3))*3
sp3=np.vstack((tp3,cp3)).T
print(len(sp3))
sp3[:20]
```
